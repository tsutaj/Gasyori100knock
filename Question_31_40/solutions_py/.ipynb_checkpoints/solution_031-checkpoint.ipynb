```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def affine(img, a, b, c, d, tx, ty, H=-1, W=-1):
    if a*d - b*c == 0:
        raise ValueError("ad - bc must not be zero")
    
    OH, OW = img.shape[0], img.shape[1]
    H, W = int(H), int(W)
    if H < 0:
        H = OH
    if W < 0:
        W = OW
        
    ch = img.shape[2]
    output_img = np.zeros((H, W, ch))
    for i in range(H):
        for j in range(W):
            x = int(1.0/(a*d-b*c) * (d*i - b*j) - tx)
            y = int(1.0/(a*d-b*c) * (a*j - c*i) - ty)
            if 0 <= x < OH and 0 <= y < OW:
                output_img[i, j] = img[x, y]
                
    output_img = np.clip(output_img, 0, 255).astype("uint8")
    return output_img
```

```python
img = cv2.imread("../imori.jpg")
H, W, ch = img.shape
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
dx, dy = 30, 30
output_img_1 = affine(img, 1, 0, dy/W, 1, 0, 0, H, W+dy)
output_img_2 = affine(img, 1, dx/H, 0, 1, 0, 0, H+dx, W)
output_img_3 = affine(img, 1, dx/H, dy/W, 1, 0, 0, H+dx, W+dy)

plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(output_img_1, cv2.COLOR_BGR2RGB))
plt.title("X-sharing")
plt.subplot(1, 3, 2)
plt.imshow(cv2.cvtColor(output_img_2, cv2.COLOR_BGR2RGB))
plt.title("Y-sharing")
plt.subplot(1, 3, 3)
plt.imshow(cv2.cvtColor(output_img_3, cv2.COLOR_BGR2RGB))
plt.title("XY-sharing")
plt.show()
```
