```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread("../imori.jpg")
H, W, ch = img.shape
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
output_img = np.zeros((H, W, ch))
for i in range(H):
    for j in range(W):
        for c in range(ch):
            s = 0
            for k in range(-1, 2, 1):
                for l in range(-1, 2, 1):
                    x = i + k
                    y = j + l
                    p = 0 if x < 0 or x >= H or y < 0 or y >= W else img[x][y][c]
                    s += p
            output_img[i][j][c] = s // 9
output_img = output_img.astype("uint8")
```

```python
plt.imshow(cv2.cvtColor(output_img, cv2.COLOR_BGR2RGB))
plt.show()
```
