```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.cvtColor(cv2.imread("../imori.jpg"), cv2.COLOR_BGR2GRAY)
H, W = img.shape
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
K = 3
pad = K // 2

output_img = np.zeros((H+2*pad, W+2*pad), dtype="uint8")
output_img[pad:pad+H, pad:pad+W] = img.copy()
tmp_img = output_img.copy()

for i in range(H):
    for j in range(W):
        ma = np.max(tmp_img[i:i+K, j:j+K])
        mi = np.min(tmp_img[i:i+K, j:j+K])
        output_img[i+pad][j+pad] = ma - mi
```

```python
plt.imshow(output_img, cmap="gray")
plt.show()
```
