```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread("../imori.jpg")
H, W, ch = img.shape
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
K = 3
pad = K // 2

output_img = np.zeros((H+pad*2, W+pad*2, ch))
output_img[pad:pad+H, pad:pad+W, :] = img.copy().astype(np.float)
tmp_img = output_img.copy()

for i in range(H):
    for j in range(W):
        for c in range(ch):
            output_img[pad+i, pad+j, c] = np.trace(tmp_img[i:i+K, j:j+K, c]) / 3
output_img = output_img[pad:pad+H, pad:pad+W, :].astype("uint8")
```

```python
plt.imshow(cv2.cvtColor(output_img, cv2.COLOR_BGR2RGB))
plt.show()
```
