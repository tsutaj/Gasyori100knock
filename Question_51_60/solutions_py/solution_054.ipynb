```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread("../imori.jpg")
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()

img_part = cv2.imread("../imori_part.jpg")
plt.imshow(cv2.cvtColor(img_part, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
def template_matching_ssd(img, img_part):
    H, W, _ = img.shape
    h, w, _ = img_part.shape
    img_f = img.copy().astype(np.float32)
    img_part_f = img_part.copy().astype(np.float32)

    min_S = np.inf
    opt_x, opt_y = -1, -1
    for i in range(H-h+1):
        for j in range(W-w+1):
            S = np.sum((img_part_f - img_f[i:i+h, j:j+w])**2)
            if min_S > S:
                min_S = S
                opt_x, opt_y = j, i
    output_img = cv2.rectangle(img.copy(), (opt_x, opt_y), (opt_x+w, opt_y+h), (0, 0, 255), 2)
    return output_img
```

```python
img_ssd = template_matching_ssd(img, img_part)
plt.imshow(cv2.cvtColor(img_ssd, cv2.COLOR_BGR2RGB))
plt.show()
```
