```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img_imori = cv2.imread("../imori.jpg")
img_thorino = cv2.imread("../thorino.jpg")
```

```python
def alpha_blend(img_1, img_2, alpha=0.6):
    img_out = img_1.astype(np.float32)*alpha + img_2.astype(np.float32)*(1-alpha)
    img_out = np.clip(img_out, 0, 255).astype("uint8")
    return img_out
```

```python
img_alpha = alpha_blend(img_imori, img_thorino, 0.6)
plt.imshow(cv2.cvtColor(img_alpha, cv2.COLOR_BGR2RGB))
plt.show()
```
