```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def grayscale(img):
    img_gray = img[:, :, 0] * 0.0722 + img[:, :, 1] * 0.7152 + img[:, :, 2] * 0.2126
    return img_gray
```

```python
img = cv2.imread("../imori.jpg")
img_binary = np.vectorize(lambda x: 0 if x < 128 else 255)(grayscale(img))
```

```python
plt.imshow(img_binary, cmap="gray")
plt.show()
```
