```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def color_reduction(value):
    if 0 <= value < 64:
        return 32
    elif 64 <= value < 128:
        return 96
    elif 128 <= value < 192:
        return 160
    elif 192 <= value < 256:
        return 224
    return -1
```

```python
img = cv2.imread("../imori.jpg")
H, W, ch = img.shape
```

```python
for i in range(H):
    for j in range(W):
        for k in range(ch):
            img[i][j][k] = color_reduction(img[i][j][k])
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```
