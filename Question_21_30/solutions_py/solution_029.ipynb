```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread("../imori.jpg")
H, W, ch = img.shape
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
tx, ty = -30, +30
scale_x, scale_y = 1.3, 0.8

output_img_1 = np.zeros((int(H*scale_y), int(W*scale_x), ch))
output_img_2 = output_img_1.copy()

H_a, W_a, _ = output_img_1.shape
for i in range(H_a):
    for j in range(W_a):
        x = int(1.0 / (scale_x * scale_y) * scale_x * i)
        y = int(1.0 / (scale_x * scale_y) * scale_y * j)
        if 0 <= x < H and 0 <= y < W: output_img_1[i, j, :] = img[x, y, :]
        x -= tx
        y -= ty
        if 0 <= x < H and 0 <= y < W: output_img_2[i, j, :] = img[x, y, :]
        

output_img_1 = output_img_1.astype("uint8")
output_img_2 = output_img_2.astype("uint8")
```

```python
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(output_img_1, cv2.COLOR_BGR2RGB))
plt.title("affine 1")
plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(output_img_2, cv2.COLOR_BGR2RGB))
plt.title("affine 2")
plt.show()
```
