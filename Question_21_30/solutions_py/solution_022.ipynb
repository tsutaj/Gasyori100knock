```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread("../imori_dark.jpg")
H, W, ch = img.shape
print(H, W, ch)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
m = np.mean(img)
s = np.std(img)
print(m, s)

m0 = 128
s0 = 52

img = s0/s * (img-m) + m0
img = np.clip(img, 0, 255).astype("uint8")
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()

plt.hist(img.flatten(), bins=255, rwidth=0.8, range=(0, 255))
```
