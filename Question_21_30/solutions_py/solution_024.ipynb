```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread("../imori_gamma.jpg")
H, W, ch = img.shape
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
c = 1
g = 2.2

m = np.max(img)
output_img = img / m
output_img = (1/c * output_img) ** (1/g)
output_img *= m

output_img = np.clip(output_img, 0, 255).astype("uint8")
```

```python
plt.imshow(cv2.cvtColor(output_img, cv2.COLOR_BGR2RGB))
plt.show()
```
