```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread("../imori.jpg")
H, W, ch = img.shape
print(H, W, ch)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
S = H*W*ch
Z = np.max(img)

cnt = 0
output_img = img.copy()
for i in range(256):
    indices = np.where(img == i)
    cnt += len(img[indices])
    output_img[indices] = Z / S * cnt
    
output_img = np.clip(output_img, 0, 255).astype("uint8")
```

```python
plt.imshow(cv2.cvtColor(output_img, cv2.COLOR_BGR2RGB))
plt.show()
```
