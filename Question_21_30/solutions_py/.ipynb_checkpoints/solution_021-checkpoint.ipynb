```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread("../imori_dark.jpg")
H, W, ch = img.shape
print(H, W, ch)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
mi = np.min(img)
ma = np.max(img)
print(mi, ma)

for c in range(ch):
    for i in range(H):
        for j in range(W):
            img[i, j, c] = (255 / (ma - mi)) * (img[i, j, c] - mi)
img = img.astype("uint8")
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```
