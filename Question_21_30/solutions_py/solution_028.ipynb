```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread("../imori.jpg")
H, W, ch = img.shape
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
tx, ty = -30, +30

output_img = np.zeros((H, W, ch))
for i in range(H):
    for j in range(W):
        ip = i + tx
        jp = j + ty
        if ip < 0 or ip >= H or jp < 0 or jp >= W:
            continue
        output_img[ip, jp] = img[i, j]

output_img = output_img.astype("uint8")
```

```python
plt.imshow(cv2.cvtColor(output_img, cv2.COLOR_BGR2RGB))
plt.show()
```
