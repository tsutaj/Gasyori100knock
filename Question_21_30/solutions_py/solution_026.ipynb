```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread("../imori.jpg")
H, W, ch = img.shape
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
scale = 1.5
H_big, W_big = int(H * scale), int(W * scale)
output_img = np.zeros((H_big, W_big, ch))

for i in range(H_big):
    for j in range(W_big):
        x, y = min(H-2, int(i/scale)), min(W-2, int(j/scale))
        dx, dy = i/scale - x, j/scale - y
        D = [(1-dx)*(1-dy), dx*(1-dy), (1-dx)*dy, dx*dy]
        I = [img[x, y, :], img[x+1, y, :], img[x, y+1, :], img[x+1, y+1, :]]
        output_img[i, j, :] = sum(d*z for (d, z) in zip(D, I))

output_img = np.clip(output_img, 0, 255).astype("uint8")
```

```python
plt.imshow(cv2.cvtColor(output_img, cv2.COLOR_BGR2RGB))
plt.show()

print(H_big, W_big, "vs", H, W)
```
